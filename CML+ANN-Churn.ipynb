```python
dataset = pd.read_csv('/content/drive/MyDrive/Churn_Modelling.csv')
x = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_x_1 = LabelEncoder()
x[:, 1] = labelencoder_x_1.fit_transform(x[:, 1])
labelencoder_x_2 = LabelEncoder()
x[:, 2] = labelencoder_x_2.fit_transform(x[:, 2])
x
```

```python
# Splitting the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
len(dataset)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python

#Classical ML

from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
pred = clf.predict(x_test)
```

```python
# Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)

print(cm)
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, pred)
print(score*100)
```

```python
# Accuracy of decision Tree classifier is around 80%
```

```python
## Random forest
#Classical ML

from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier(n_estimators=200) #n_estimators means the number of decision trees
clf2.fit(x_train, y_train)
pred2 = clf2.predict(x_test)
```

```python
# Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(y_test, pred2)

print(cm2)
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score

score2 = accuracy_score(y_test, pred2)
print(score2*100)
```

```python
# Accuracy score for random forest is 87%
```

```python
# Now let's make the ANN
# Importing the Keras Libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
#units: It's an art, comes by experience, input+
#initializer: how your weights are updated
#relu activation function
#input_dim: input shape (shape of the data, how many features do we have in the data)
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))

#Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the third hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding Output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
len(x_train)
```

```python
# Compile the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
y_train[0]
# Fitting the ANN to the Training set
classifier.fit(x_train, y_train, batch_size =10, epochs = 50)
```

```python
# Part 3 - Making predictions and evaluating the model
# predicting the Test set results
y_pred_ann = classifier.predict(x_test)
y_pred_ann = (y_pred_ann > 0.5)

cm3 = confusion_matrix(y_test, y_pred_ann)
print(cm3)
```

```python
# Making the confusion Matrix
from sklearn.metrics import accuracy_score

score3 = accuracy_score(y_test,y_pred_ann)
print(score3*100)
```
